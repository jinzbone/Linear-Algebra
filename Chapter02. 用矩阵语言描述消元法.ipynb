## 用矩阵语言描述消元法

用矩阵语言描述消元法（当方程组有解时，消元法就能有效得出一个解）  
核心：矩阵变换 <br>
求解线性方程组
$\begin{align}
&x+2y+z=2\\
&3x+8y+z=12\\
&4y+z=2
\end{align}$

### 1. 消元法 Elimination
<br>
$\left[
\begin{matrix}
   \color{red}1 & 2 & 1 \\
   3 & 8 & 1\\
   0 & 4 & 1\\
  \end{matrix}
  \right]$
$ \xrightarrow[]{消元} $
$\left[
\begin{matrix}
   \color{red}1 & 2 & 1\\
   0 & \color{red}2 & -2\\
   0 & 4 & 1\\
  \end{matrix}
  \right]$
$ \xrightarrow[]{消元} $
$\left[
\begin{matrix}
    \color{red}1&2&1\\
    0&\color{red}2&-2\\
    0&0&\color{red}5\\
    \end{matrix}
    \right]$ 
<br>
<br>
$\qquad A\qquad \qquad \qquad \longrightarrow \qquad \qquad \qquad \qquad  U$  
<br>
$\color{red}{1\quad 2 \quad 5}$ 表示主元

### 2. 回代 Back Substitution

$\left[
\begin{matrix}
   \color{red}1 & 2 & 1&2 \\
   3 & 8 & 1&12\\
   0 & 4 & 1&2\\
  \end{matrix}
  \right]$
$\xrightarrow[]{消元}$
$\left[
\begin{matrix}
   \color{red}1 & 2 & 1&2\\
   0 & \color{red}2 & -2&6\\
   0 & 4 & 1&2\\
  \end{matrix}
  \right]$
$ \xrightarrow[]{消元} $
$\left[
\begin{matrix}
    \color{red}1&2&1&2\\
    0&\color{red}2&-2&6\\
    0&0&\color{red}5&-10\\
    \end{matrix}
    \right]$ 
$\dashrightarrow$
$\begin{align}
&x+2y+z=2\\
&2y-2z=6\\
&5z=-10
\end{align}$
$\xrightarrow[]{回代}$
$\begin{align}
&x=2\\
&y=1\\
&z=-2
\end{align}$
<br>
<br>
$\quad\quad A \qquad\ \ b \ \ \qquad\qquad\qquad\qquad\qquad\qquad \quad \ U \qquad\quad\ \ c $     
$\qquad\ \ \ 增广矩阵$
